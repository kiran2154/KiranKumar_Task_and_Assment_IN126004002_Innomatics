```python
def parking_lot(capacity,logs):
  currently_parked=0
  peak_parking=0
  #proceesing
  for log in logs:
    if log == 'IN':
      currently_parked += 1
    elif log == 'OUT':
      currently_parked -= 1
    if currently_parked > peak_parking:
      peak_parking = currently_parked
  print("Currently Parked Vehicles: ",currently_parked)
  if currently_parked > capacity:
            print("Parking Status: Capacity Exceeded")
  else:
        print("Parking Status: Available")
capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]
parking_lot(capacity, vehicle_logs)
```

```python
def estimate_delivery_time(distance, traffic_level, weather):
    # Assume Base time calculation (5 minutes per km)
    base_time = distance * 5
    # Traffic adjustment
    if traffic_level.lower() == "low":
        traffic_delay = 5
    elif traffic_level.lower() == "medium":
        traffic_delay = 10
    else:  # High
        traffic_delay = 15
    # Weather adjustment
    if weather.lower() == "clear":
        weather_delay = 0
    elif weather.lower() == "cloudy":
        weather_delay = 5
    else: # rainy
        weather_delay = 10
    final_time = base_time + traffic_delay + weather_delay

    print("Estimated Delivery Time:", final_time, "minutes")
estimate_delivery_time(8, "High", "Rainy")
```

```python
def seat_booking(booked_seats,total_seats):
  booked_count = 0
  for seat in booked_seats:
    if seat == 1:
      booked_count += 1
  available_seats = total_seats - booked_count
  occupancy = (booked_count/total_seats) * 100
  print("Occupancy:",int(occupancy),"%")
  if occupancy == 100:
        status = "Housefull"
  elif occupancy >= 75:
        status = "Almost Full"
  else:
        status = "Seats Available"
  print("Show Status:", status)
  if occupancy >= 90:
        print("No extra show needed.")
  return occupancy, status
total_seats = 200
booked_seats = [1] * 150   # 150 seats booked
occupancy, status = seat_booking(booked_seats, total_seats)
```

```python
def classify_server_load(cpu_readings):
  total = 0
  for reading in cpu_readings:
    total += reading
  average = total / len(cpu_readings)
  print("Average CPU Load:", int(average), "%")
  if average < 50:
        status = "Normal"
  elif average <= 80:
        status = "Warning"
  else:
        status = "Critical"
  print("Server Status:", status)
cpu_readings = [45, 60, 70, 85, 90]
classify_server_load(cpu_readings)
```

```python
def track_classroom_usage(resource_usage):
  overused_resource = []
  for resource, hours in resource_usage.items():
    # Check each resource
    if hours > 8:
      overused_resource.append(resource)
  if overused_resource:
      print("Overused Resources:", ", ".join(overused_resource))
      print("Energy Alert: Yes")
  else:
      print("Overused Resources: None")
      print("Energy Alert: No")
resource_usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}
track_classroom_usage(resource_usage)
```

```python
def manage_event_registration(capacity, total_registrations):
    confirmed = 0
    waitlisted = 0
    for i in range(total_registrations):
        if confirmed < capacity:
            confirmed += 1
        else:
            waitlisted += 1
    # status
    if confirmed >= capacity:
        status = "Closed"
    else:
        status = "Open"
    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)
capacity = 100
registrations = 105

manage_event_registration(capacity, registrations)
```
