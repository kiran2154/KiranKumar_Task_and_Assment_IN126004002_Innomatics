```python
class Smartlight:
  def __init__(self, name):
    self.name = name
    self.status = "OFF"
  def turn_on(self):
    self.status = "ON"
  def turn_off(self):
    self.status = "OFF"
  def display_status(self):
    print(f"{self.name} is {self.status}")
  #INPUT
light_name = input("Light Name: ")
action = input("Action: ").upper()
light = Smartlight(light_name)
if action == "ON":
    light.turn_on()
elif action == "OFF":
    light.turn_off()
light.display_status()  #DISPLAY
```

```python
class Employee:
  def __init__(self, name,emp_id,department):
    self.name = name
    self.emp_id = emp_id
    self.department = department
  def display_idcard(self):
    print("Name: ",self.name)
    print("Employee ID: ",self.emp_id)
    print("department: ",self.department)
  #INPUT
name = input("Employee Name: ")
emp_id = input("Employee ID: ")
department = input("Department: ")
emp = Employee(name, emp_id, department)  #OBJECT
emp.display_idcard()  #DISPLAY
```

```python
class Phone:
  def __init__(self,name,phoneno):
    self.name=name
    self.phoneno=phoneno
  def display_info(self):
    print("Contact Saved")
    print("Contact Name: ",self.name)
    print("Phone Number: ",self.phoneno)
  #INPUT
name = input("Contact Name: ")
phoneno = input("Phone Number: ")
ph = Phone(name, phoneno)  #OBJECT
ph.display_info()  #DISPLAY
```

```python
class Product:
  def __init__(self,name,price):
    self.name=name
    self.price=price
  def display_info(self):
    print("Product Name: ",self.name)
    print("Price: ₹",self.price)
#INPUT
name = input("Product Name: ")
price = int(input("Price: "))
prod = Product(name, price)  #OBJECT
prod.display_info()  #DISPLAY
```

```python
class Movies:
  def __init__(self,name,rating):
    self.name=name
    self.rating=rating
  def display_info(self):
    print("Movie Name: ",self.name)
    print(f"Rating: {self.rating}/5")
#INPUT
name = input("Movie Name: ")
rating = float(input("Rating: "))
mov=Movies(name,rating)  #OBJECT
mov.display_info()  #DISPLAY
```

```python
class Delivery:
  def __init__(self,name,address):
    self.name=name
    self.address=address
  def display_info(self):
    print("Delivery Details")
    print("Customer: ",self.name)
    print("Address: ",self.address)
#INPUT
name = input("Customer Name: ")
address = input("Address: ")
delv=Delivery(name,address)  #OBJECT
delv.display_info()  #DISPLAY
```
